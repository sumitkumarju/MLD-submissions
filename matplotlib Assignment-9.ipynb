```python
import numpy as np
```

```python
Max = [39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]
Min = [21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]
month = np.linspace(1, 12, num=12)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x=month,y=Max)
plt.scatter(x=month,y=Min)
```

```python
from scipy import optimize

def test_func(x, a, b,c):
    return a +b* np.sin(c*x)

params, params_covariance = optimize.curve_fit(test_func, month, Max,
                                               p0=[40, 10,1/4])

print(params)
```

```python
params1, params_covariance = optimize.curve_fit(test_func, month, Min,
                                               p0=[20, 15,1/4])
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(month, Max, label='Data1')
plt.plot(month, test_func(month, params[0], params[1],params[2]),
         label='Fitted function1')
plt.scatter(month, Min, label='Data2')
plt.plot(month, test_func(month, params1[0], params1[1],params1[2]),
         label='Fitted function2')

plt.legend(loc='best')

plt.show()
```
